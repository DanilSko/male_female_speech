```python
from os import walk
import requests
import re
import codecs
folder_with_plays = '/Users/daniilskorinkin/Documents/GitHub/rusdracor/tei'
```

```python
def record_and_close_file (outfile, outspeech):
    outfile.write(outspeech)
    outfile.close()
```

```python
def write_speech (playname):
    male_speech = get_speech (playname, 'MALE')
    female_speech = get_speech (playname, 'FEMALE')
    outmale = codecs.open ('MALE/' + 'MALE_'+ playname+'.txt', 'w', 'utf-8')
    outfemale = codecs.open ('FEMALE/' + 'FEMALE_'+ playname+'.txt', 'w', 'utf-8')
    record_and_close_file (outmale, male_speech)
    record_and_close_file (outfemale, female_speech)
```

```python
def get_speech (playname, gender):
    link = 'https://dracor.org/api/corpora/rus/play/{}/spoken-text?gender={}'
    link = link.format(playname,gender)
    print (link)
    text = requests.get(link).text
    return (text)
```

```python
for path, dirs, filenames in walk(folder_with_plays):
    for filename in filenames:
        if '.xml' in filename:
            playname = re.sub ('.xml', '', filename)
            write_speech (playname)
```
